```python
# Figure 2.3
f22a = '/data/keeling/a/jamessg3/Data/Figures/keax_20190525_0059_SRV_3.1.png'
f22b = '/data/keeling/a/jamessg3/Data/Figures/keax_20190525_0110_SRV_0.5.png'



im22a = plt.imread(f22a)
im22b = plt.imread(f22b)
```

```python
%matplotlib widget

fig, axs = plt.subplot_mosaic([['a)', 'b)']], constrained_layout=True, figsize=(12,6))#, subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(16,6))
fig.gca().set_xticks([]);
fig.gca().set_yticks([]);
props = dict(boxstyle='round', facecolor='white', alpha=0.8)

for label, ax in axs.items():
    ax.tick_params(
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        top=False, 
        left=False, # ticks along the top edge are off
        labelbottom=False,
        labelleft= False) # labels along the bottom edge are off
    ax.set_title('', fontstyle='italic')
    ax.set_title(label, fontfamily='sans_serif', loc='left', fontsize=16)
    

axs['a)'].imshow(im22a)
s1 = 310
s2 = 290
l1 = 16
l2 = -15
axs['a)'].arrow(s1,s2,l1,l2, width = 5, color = 'white', shape = 'full')
axs['a)'].arrow(s1+l1,s2+l2, -l1,-l2, width = 5, color = 'white', shape = 'full')
axs['a)'].text(370, 245, '1.0 km', fontsize=14, verticalalignment='center', bbox=props)
axs['b)'].imshow(im22b)

s1 = 610
s2 = 280
l1 = 35
l2 = 70
axs['b)'].arrow(s1,s2,l1,l2, width = 5, color = 'white', shape = 'full')
axs['b)'].arrow(s1+l1,s2+l2, -l1,-l2, width = 5, color = 'white', shape = 'full')
s1 = 671
s2 = 425
l1 = 10
l2 = 100
axs['b)'].arrow(s1,s2, l1, l2, width = 5, color = 'white', shape = 'full')
axs['b)'].arrow(s1+l1,s2+l2, -l1,-l2, width = 5, color = 'white', shape = 'full')

axs['b)'].text(670, 280, '7.25 km', fontsize=14, verticalalignment='center', bbox=props)
axs['b)'].text(720, 475, '7.5 km', fontsize=14, verticalalignment='center', bbox=props)
fig.savefig("../Figures/fig2d3.png", bbox_inches='tight', dpi = 700)
```
